```python
#Librerias
import pandas as pd
import winsound
import time
import gensim 
from gensim.models.ldamulticore import LdaMulticore
from gensim import corpora, models
import pyLDAvis.gensim #LDA visualization library

from nltk.corpus import stopwords
import string
from nltk.stem.wordnet import WordNetLemmatizer

import warnings
warnings.simplefilter('ignore')
from itertools import chain

import numpy as np
from textblob import TextBlob
```

```python
#Sound function
def beep():
    duration = 60000
    freq = 3000  # Hz
    winsound.Beep(freq, duration)
```

```python
#Lectura de datos
datafile1 = 'D:\Downloads\Iphone.csv'
datafile2 = 'D:\Downloads\Samsung.csv'
datafile3 = 'D:\Downloads\Huawei.csv'
datafile4 = 'D:\Downloads\Oppo.csv'
datafile5 = 'D:\Downloads\Xiaomi.csv'


#raw_data = pd.read_csv(datafile, parse_dates=[0], infer_datetime_format=True)
data=[]
data.append(pd.read_csv(datafile1))
data.append(pd.read_csv(datafile2))
data.append(pd.read_csv(datafile3))
data.append(pd.read_csv(datafile4))
data.append(pd.read_csv(datafile5))

data1 = pd.read_csv(datafile1)
data2 = pd.read_csv(datafile2)
data3 = pd.read_csv(datafile3)
data4 = pd.read_csv(datafile4)
data5 = pd.read_csv(datafile5)


#Adding camps
data1['Polarity']=""
data1["Subjetivity"]=""
data1['Topic']=""
data2['Polarity']=""
data2["Subjetivity"]=""
data2['Topic']=""
data3['Polarity']=""
data3["Subjetivity"]=""
data3['Topic']=""
data4['Polarity']=""
data4["Subjetivity"]=""
data4['Topic']=""
data5['Polarity']=""
data5["Subjetivity"]=""
data5['Topic']=""


dataText=[]
for i in range (0,5):
    dataText.append(data[i]['Text'])

```

```python
#Limpieza de datos
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()
def clean(text):
    stop_free = ' '.join([word for word in text.lower().split() if word not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = ' '.join([lemma.lemmatize(word) for word in punc_free.split()])
    return punc_free.split()

data1['Clean']=data1['Text'].apply(clean)
data2['Clean']=data2['Text'].apply(clean)
data3['Clean']=data3['Text'].apply(clean)
data4['Clean']=data4['Text'].apply(clean)
data5['Clean']=data5['Text'].apply(clean)
```

```python
#Diccionario de palabras
dictionary1 = corpora.Dictionary(data1['Clean'])
dictionary2 = corpora.Dictionary(data2['Clean'])
dictionary3 = corpora.Dictionary(data3['Clean'])
dictionary4 = corpora.Dictionary(data4['Clean'])
dictionary5 = corpora.Dictionary(data5['Clean'])

```

```python
#Matriz de terminos
doc_term_matrix1 = [dictionary1.doc2bow(doc) for doc in data1['Clean'] ]
doc_term_matrix2 = [dictionary2.doc2bow(doc) for doc in data2['Clean'] ]
doc_term_matrix3 = [dictionary3.doc2bow(doc) for doc in data3['Clean'] ]
doc_term_matrix4 = [dictionary4.doc2bow(doc) for doc in data4['Clean'] ]
doc_term_matrix5 = [dictionary5.doc2bow(doc) for doc in data5['Clean'] ]
```

```python
#Instancia del modelo
lda = gensim.models.ldamodel.LdaModel
```

```python
#Modelado de palabras topicas
%time ldamodel1 = LdaMulticore(doc_term_matrix1,num_topics=10,id2word=dictionary1,passes=100,minimum_probability=0)
%time ldamodel2 = LdaMulticore(doc_term_matrix2,num_topics=10,id2word=dictionary2,passes=100,minimum_probability=0)
%time ldamodel3 = LdaMulticore(doc_term_matrix3,num_topics=10,id2word=dictionary3,passes=100,minimum_probability=0)
%time ldamodel4 = LdaMulticore(doc_term_matrix4,num_topics=10,id2word=dictionary4,passes=100,minimum_probability=0)
%time ldamodel5 = LdaMulticore(doc_term_matrix5,num_topics=10,id2word=dictionary5,passes=100,minimum_probability=0)
```

```python
# Assigns the topics to the documents in corpus
start_time = time.time()
lda_corpus1 = ldamodel1[doc_term_matrix1]
lda_corpus2 = ldamodel2[doc_term_matrix2]
lda_corpus3 = ldamodel3[doc_term_matrix3]
lda_corpus4 = ldamodel4[doc_term_matrix4]
lda_corpus5 = ldamodel5[doc_term_matrix5]

for j in range (0,len(data1['Topic'])):
    val1=[]
    for i in range (0,9):
        val1.append([lda_corpus1[j][i][1],lda_corpus1[j][i][0]])
    d=max(val1)
    data1['Topic'][j]=dic1[d[1]]

for j in range (0,len(data2['Topic'])):
    val2=[]
    for i in range (0,9):
        val2.append([lda_corpus2[j][i][1],lda_corpus2[j][i][0]])
    d=max(val2)
    data2['Topic'][j]=dic2[d[1]]

for j in range (0,len(data3['Topic'])):
    val3=[]
    for i in range (0,9):
        val3.append([lda_corpus3[j][i][1],lda_corpus3[j][i][0]])
    d=max(val3)
    data3['Topic'][j]=dic3[d[1]]

for j in range (0,len(data4['Topic'])):
    val4=[]
    for i in range (0,9):
        val4.append([lda_corpus4[j][i][1],lda_corpus4[j][i][0]])
    d=max(val4)
    data4['Topic'][j]=dic4[d[1]]

for j in range (0,len(data5['Topic'])):
    val5=[]
    for i in range (0,9):
        val5.append([lda_corpus5[j][i][1],lda_corpus5[j][i][0]])
    d=max(val5)
    data5['Topic'][j]=dic5[d[1]]
    
elapsed_time = time.time() - start_time
print(elapsed_time/60)
```

```python
#Deleting no comercial tweets
data1 = data1[data1.Topic != 'no commercial']
data2 = data2[data2.Topic != 'no commercial']
data3 = data3[data3.Topic != 'no commercial']
data4 = data4[data4.Topic != 'no commercial']
data5 = data5[data5.Topic != 'no commercial']

```

```python
#Adding analysis
start_time = time.time()

for i in range (0,len(dataText[0])):
    blob = TextBlob(dataText[0][i])
    data1["Polarity"][i]=blob.sentiment.polarity
    data1["Subjetivity"][i]=blob.sentiment.subjectivity
for i in range (0,len(dataText[1])):
        blob = TextBlob(dataText[1][i])
        data2["Polarity"][i]=blob.sentiment.polarity
        data2["Subjetivity"][i]=blob.sentiment.subjectivity
for i in range (0,len(dataText[2])):
        blob = TextBlob(dataText[2][i])
        data3["Polarity"][i]=blob.sentiment.polarity
        data3["Subjetivity"][i]=blob.sentiment.subjectivity
for i in range (0,len(dataText[3])):
        blob = TextBlob(dataText[3][i])
        data4["Polarity"][i]=blob.sentiment.polarity
        data4["Subjetivity"][i]=blob.sentiment.subjectivity
for i in range (0,len(dataText[4])):
        blob = TextBlob(dataText[4][i])
        data5["Polarity"][i]=blob.sentiment.polarity
        data5["Subjetivity"][i]=blob.sentiment.subjectivity
elapsed_time = time.time() - start_time
print(elapsed_time/60)
```
